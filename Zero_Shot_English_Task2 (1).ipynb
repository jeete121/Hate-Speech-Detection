```
from datasets import load_dataset, load_metric
import numpy as np
import math
from sklearn.metrics import classification_report
```

```
dataset1['task2'].value_counts()
```

```
candidate_labels = ["NONE", "PRFN","OFFN","HATE"]
predictions = []
for offset in tqdm(range(math.ceil(len(dataset1)/100))):
    preds = classifier([dataset1["text"][100*offset+i] for i in range(100) if 100*offset+i<len(dataset1)], candidate_labels)
    pred_labels = [pred["labels"][np.argmax(pred["scores"])] for pred in preds]
    predictions.extend([candidate_labels.index(pred_label) for pred_label in pred_labels])
```
